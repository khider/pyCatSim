Cat adoption day!

Let's go to the adoption center and meet some cats!

```python
import pyCatSim as cats

nutmeg = cats.Cat(name='Nutmeg', age = 3, color = 'tortoiseshell')
nutmeg.show()
```

```python
nutmeg.mood
```

```python
nutmeg.play()
nutmeg.mood
```

Ok, let's adopt Nutmeg!

```python
Deborah = cats.Owner(name="Sasha", cats_owned=[])
```

```python
Deborah.adopt(nutmeg)
```

```python
Deborah.cats_owned[0].name
```

Upon coming back home, Nutmeg is very hungry:

```python
Deborah.feed(nutmeg)
nutmeg.mood
```

Deborah learns a lot about random cat facts:

```python
Deborah.give_fact()
```

Nutmeg is very happy to be adopted and purrs!

```python
nutmeg.make_noise(noise='purr', play = True)
```
